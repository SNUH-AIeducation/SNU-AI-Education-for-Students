# Decision Tree 실습

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree
```

```
# import example data
data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df["class"] = data.target
df['class'] = df['class'].map(dict(zip([0,1,2],data.target_names)))
df
```

```
# split dataset
df = df.sample(frac=1)

x_data = df.drop(columns=['class'])
y_data = df['class']

ratio = 1/10
idx = int(x_data.shape[0]*(1-ratio))

x_train = x_data.iloc[:idx]
x_test  = x_data.iloc[idx:]
y_train = y_data.iloc[:idx]
y_test  = y_data.iloc[idx:]
```

```
# train model
model = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4,min_samples_leaf=4)
model.fit(x_train, y_train)
```

```
# model test
print('Train Accuracy :', np.round(model.score(x_train, y_train),4))
print('Test Accuracy :\t', np.round(model.score(x_test, y_test),4))
```

```
# plot decision tree
plt.figure(figsize=(12,12))
tree.plot_tree(model,feature_names=df.columns[:4])
plt.show()
```

```
# plot feature importance
plt.figure(figsize=(8,6))
plt.bar(height=model.feature_importances_, x=df.columns[:4])
plt.show()
```

# Random Forest 실습

```
from sklearn.ensemble import RandomForestClassifier
```

```
model = RandomForestClassifier(criterion='entropy', max_depth=4,min_samples_leaf=4)
model.fit(x_train, y_train)
```

```
# model test
print('Train Accuracy :', np.round(model.score(x_train, y_train),4))
print('Test Accuracy :\t', np.round(model.score(x_test, y_test),4))
```

```
# plot feature importance
plt.figure(figsize=(8,6))
plt.bar(height=model.feature_importances_, x=df.columns[:4])
plt.show()
```
